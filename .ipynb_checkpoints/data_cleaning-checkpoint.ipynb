# data manipulation
- hour, weekdays, month extraction
- integration with weather

```python
hourly_data=pd.read_csv("/Users/jchen056/Data/MTA_Subway_Hourly_Ridership__Beginning_February_2022.csv")
hourly_data.info()
```

```python
hourly_data["datetime"]=hourly_data["transit_timestamp"].apply(lambda x: datetime.strptime(x,"%m/%d/%Y %I:%M:%S %p"))
hourly_data.head()
```

```python
hourly_data["hour"]=hourly_data["datetime"].apply(lambda x: x.hour)
hourly_data['month']=hourly_data["datetime"].apply(lambda x: x.month)
hourly_data["weekdays"]=hourly_data["datetime"].apply(lambda x: x.weekday())
```

```python
def extract_date(x):
    return datetime(x.year,x.month,x.day)

hourly_data['date']=hourly_data['datetime'].apply(extract_date)
```

```python
# hourly_data.routes.value_counts()
def count_num_routes(routes):
    routes=str(routes)
    if routes=="SIR":
        return 1
    return len(''.join([i for i in routes.strip() if i!=',']))

hourly_data['num_lines']=hourly_data['routes'].apply(count_num_routes)
```

```python
# # hourly_data.station_complex.unique()
# hourly_data[hourly_data['station_complex']=='Times Sq-42 St (N,Q,R,W,S,1,2,3,7)/42 St (A,C,E)/Bryant Pk (B,D,F,M)/5 Av (7)']
```

```python
weather=pd.read_csv("Data/weather.csv")
weather['Date']=weather['DATE'].apply(lambda x: datetime.strptime(x,"%Y-%m-%d"))
weather.head(2)
```

```python
hrly_data_with_weather=pd.merge(hourly_data,weather,how='left',left_on='date',right_on='Date')
hrly_data_with_weather.columns
```

```python
data=hrly_data_with_weather[['station_complex_id', 'station_complex', 'borough',
                       'payment_method', 'ridership','latitude','longitude',
                       'hour', 'month','weekdays', 'num_lines', 'date',
                        'TMAX', 'TMIN','PRCP'
                       ]]
data.head(2)
```

```python
data['avgT']=(data['TMAX']+data['TMIN'])/2

data_ridership=data.groupby(['station_complex','borough',
              'date','hour','month','weekdays','num_lines'])['ridership'].sum().reset_index()
```

```python
data_ridership.columns
```

```python
data_ridership_w_weather=pd.merge(data_ridership,weather,how='left',left_on='date',right_on='Date')[[
    'station_complex', 'borough', 'date', 'hour', 'month', 'weekdays',
       'num_lines', 'ridership',
                        'TMAX', 'TMIN','PRCP'
                       ]]
data_ridership_w_weather['avg_T']=(data_ridership_w_weather['TMAX']+data_ridership_w_weather['TMIN'])/2
data_ridership_w_weather.head(5)
```

```python
data_ridership_w_weather.to_csv("hourly_data_aggregated.csv")
```
