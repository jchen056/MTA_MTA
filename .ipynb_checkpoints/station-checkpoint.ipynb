```python
def daily_monthly_aggregation(hourly_data,station_name):
    station=hourly_data[hourly_data['station_complex']
    ==station_name]
    # station['date'].value_counts().plot(kind='bar')
    station_hrly=station.groupby('datetime')['ridership'].sum().reset_index()

    station_hrly['date']=station_hrly['datetime'].apply(lambda x: datetime(x.year, x.month, x.day))
    station_hrly['month']=station_hrly['datetime'].apply(lambda x: datetime(x.year,x.month,1))
    station_daily=station_hrly.groupby('date')['ridership'].agg(['sum','mean','std','size'])[:-3]
    station_daily.reset_index(inplace=True)

    station_monthly=station_hrly.groupby('month')['ridership'].agg(['sum','mean','std','size'])[:-1]
    return station1_daily,station1_monthly

```

```python
#save station hourly data to csv
def saving_hrly_station_data(hourly_data,station_name):
    station=hourly_data[hourly_data['station_complex']
    ==station_name]
    # station['date'].value_counts().plot(kind='bar')
    station_hrly=station.groupby('datetime')['ridership'].sum().reset_index()

    station_hrly['date']=station_hrly['datetime'].apply(lambda x: datetime(x.year, x.month, x.day))
    station_hrly['month']=station_hrly['datetime'].apply(lambda x: datetime(x.year,x.month,1))
    
    station_name_shortened=station_name.split('(')[0].strip()
    station_hrly.to_csv('Data/'+station_name_shortened+'.csv')
saving_hrly_station_data(hourly_data,'Grand Central-42 St (S,4,5,6,7)')
```

```python
def station_vis(station_name):
    station_name_s=station_name.split('(')[0].strip()
    station_hrly=pd.read_csv('Data/'+station_name_s+'.csv')
    station_hrly=station_hrly[station_hrly['month']!='2023-10-01']

    station_daily=station_hrly.groupby('date')['ridership'].agg(['sum','mean','std','size'])
    station_daily.reset_index(inplace=True)

    station_monthly=station_hrly.groupby('month')['ridership'].agg(['sum','mean','std','size'])
    station_monthly.reset_index(inplace=True)
#     print(station_monthly.iloc[-1])

    fig=make_subplots(rows=2,cols=2,
                     subplot_titles=("Daily Ridership Data Count","Total Daily Ridership",
                                    "Monthly Ridership Data Count","Total Monthly Ridership"))

    fig.add_trace(go.Bar(x=station_daily.date,
                        y=station_daily["size"],
                        showlegend=False),row=1,col=1
                 )
    fig.add_trace(go.Bar(x=station_monthly.month,
                        y=station_monthly["size"],
                        showlegend=False),row=2,col=1)

    fig.add_trace(go.Bar(x=station_daily.date,
                         y=station_daily['sum'],
                        showlegend=False),
                 row=1,col=2)
    fig.add_trace(go.Bar(x=station_monthly.month,
                         y=station_monthly['sum'],
                        showlegend=False),
                 row=2,col=2)
    fig.update_layout(height=1000, width=1000,title_text=station_name_s+' Ridership Time Series Analysis(Feb 2022-Sep 2023)')
    fig.write_html("visualizations/station/"+station_name_s+'.html')
```

```python
for i in hourly_data['station_complex'].unique():
    station_vis(i)
```
